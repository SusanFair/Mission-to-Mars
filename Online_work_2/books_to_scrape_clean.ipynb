```python
# Import dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# Visit the Quotes to Scrape the Site http://quotes.toscrape.com/
url = 'http://books.toscrape.com/'
browser.visit(url)

# Parse the HTML
html = browser.html
book_soup = soup(html, 'html.parser')
```

```python
nav_list = book_soup.find('ul', class_='nav-list')
nav_list
```

```python
li_list = nav_list.find('li')
li_list
```

```python
a_list= li_list.find_all('a')
print(a_list)
```

```python
genres = [a.text.strip() for a in a_list]
genres
```
